# init_temperatures

```python
import numpy as np
from optimize import optimize_sa2, energy
from matplotlib import pyplot as plt

K = 4
times = 40
test_temps = [150 * i for i in range(times)]
result = [0 for _ in range(times)]

for i in range(times):
    result[i] = optimize_sa2(K, initial_temp=test_temps[i], ws=25, wc=1)
plt.xlabel("init_temperature")
plt.ylabel("contrast")
plt.plot(test_temps, list(map(lambda x: x.contrast, result)), label="contrast")
plt.plot(test_temps, list(map(lambda x: energy(x.variables, ws=25, wc=1), result)), label="energy", color="red")
plt.legend()
```

```python
plt.xlabel("init_temperature")
plt.ylabel("energy")
plt.plot(test_temps, list(map(lambda x: energy(x.variables, ws=25, wc=1), result)), label="energy", color="red")
```
