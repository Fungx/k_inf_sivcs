# init_temperatures

```python
import numpy as np
from optimize import optimize_sa1
from matplotlib import pyplot as plt

K = 4
times = 40
test_temps = [150 * i for i in range(times)]
result = [0 for _ in range(times)]

for i in range(times):
    result[i] = optimize_sa1(K, initial_temp=test_temps[i])
plt.xlabel("init_temperature")
plt.ylabel("contrast")
plt.plot(test_temps, list(map(lambda x: x.contrast, result)))
```

# ws: safety property

```python
times = 150
test_ws = [0.25 * i for i in range(times)]
result = [optimize_sa1(K, initial_temp=2500, ws=test_ws[i]) for i in range(times)]

plt.xlabel("ws")
plt.ylabel("contrast")
plt.plot(test_ws, list(map(lambda x: x.contrast, result)))
```
